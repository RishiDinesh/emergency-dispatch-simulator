```python
import openai
import os
import wave


BOSON_API_KEY = os.getenv("API_KEY")

client = openai.Client(
    api_key=BOSON_API_KEY,
    base_url="https://hackathon.boson.ai/v1"
)

# for this api, we onlu support PCM format output
response = client.audio.speech.create(
    model="higgs-audio-generation-Hackathon",
    voice="en_man",
    instructions = "crying hysterically",
    input="[SPEAKER 0] Je ne sais pas quoi faire, aidez-moi s'il vous plaît",
    response_format="pcm",
    speed = 2.0
)

# You can use these parameters to write PCM data to a WAV file
num_channels = 1        
sample_width = 2        
sample_rate = 24000   

pcm_data = response.content

with wave.open('belinda_test.wav', 'wb') as wav:
    wav.setnchannels(num_channels)
    wav.setsampwidth(sample_width)
    wav.setframerate(sample_rate)
    wav.writeframes(pcm_data)
```

```python
from openai import OpenAI
import base64
import os

BOSON_API_KEY = os.getenv("API_KEY")
client = OpenAI(api_key=BOSON_API_KEY, base_url="https://hackathon.boson.ai/v1")

def b64(path):
    return base64.b64encode(open(path, "rb").read()).decode("utf-8")

reference_path1 = "./emotion_sample.wav"
reference_transcript1 = "[SPEAKER1] Kids are talking by the door"

reference_path2 = "./emotion_sample3.wav"
reference_transcript2 = "[SPEAKER1] Dogs are sitting by the door"

resp = client.chat.completions.create(
    model="higgs-audio-generation-Hackathon",
    messages=[
        {"role": "user", "content": reference_transcript1},
        {
            "role": "assistant",
            "content": [{
                "type": "input_audio",
                "input_audio": {"data": b64(reference_path1), "format": "wav"}
            }],
        },
        {"role": "user", "content": reference_transcript2},
        {
            "role": "assistant",
            "content": [{
                "type": "input_audio",
                "input_audio": {"data": b64(reference_path2), "format": "wav"}
            }],
        },
        {"role": "user", "content": f"[SPEAKER1] The gun just went off and he... he shot himself in the face!"}
    ],
    modalities=["audio"],
    audio = {"format": "wav"},
    max_completion_tokens=4096,
    temperature=1.0,
    stream=True,
)

# audio_b64 = resp.choices[0].message.audio.data
# open("output.wav", "wb").write(base64.b64decode(audio_b64))
```

```python
aud_bytes_list = []

for chunk in aud_chunks:
    # make sure it's str, strip whitespace/newlines
    s = chunk.strip().replace("\n", "").replace(" ", "")
    # fix missing padding if needed
    m = len(s) % 4
    if m:
        s += "=" * (4 - m)
    # decode each base64 string -> raw bytes
    aud_bytes_list.append(base64.b64decode(s))

# join all decoded bytes
full_audio_bytes = b"".join(aud_bytes_list)

# optional: get base64-encoded string again if needed
aud_string = base64.b64encode(full_audio_bytes).decode("ascii")

# finally, write to WAV file
with open("output.wav", "wb") as f:
    f.write(full_audio_bytes)
```
